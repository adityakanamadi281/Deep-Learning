```python
 import tensorflow as tf
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["variety"]=le.fit_transform(df["variety"])
df

# It is a conversion of categorical to numerical 
```

```python
x=df.drop("variety",axis=1)
x
```

```python
y=df["variety"]
y
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 classes
])

# softmax divide the entire output 
```

```python
model
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# adam : adaptive learning rate 
```

```python
# Train the model
#if you have 80 samples,batch size=10.80/10,
#model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
model.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=16,              # How many samples per update
    validation_data=(x_test, y_test),
    verbose=1                   # 0 = silent, 1 = progress bar, 2 = one line per epoch
)

# epoches : It is one of the complete pass through the entire dataset by the model 
#  one epoche is not enough to train the the model 
```

```python
# # Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {accuracy:.2f}')
```
